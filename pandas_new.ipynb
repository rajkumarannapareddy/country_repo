```python
df=pd.read_csv("C:\\Users\\RAJKUMAR A\\Downloads\\countries.csv")
print(df)
```

```python
df["country"].value_counts()
```

```python
population_by_continent = df.groupby("continent")["population"].sum()
print(population_by_continent)
```

```python
population_by_country = df[df["country"].isin(["India","Bahrain","Austria"])].groupby("country")["population"].sum()
print(population_by_country)
```

```python
population_by_country = df[df["country"].isin(["India", "Benin", "Belgium", "Austria"])].groupby("country")["lifeExpectancy"].sum()
                          
sorted_values=population_by_country.sort_values(ascending=False)

print(sorted_values)
```

```python
population_top_countries=df.groupby("country")["population"].sum()
print(population_top_countries) 
```
